```python
# Create New Year column for each df
df_2015['year'] = '2015'
df_2016['year'] = '2016'
df_2017['year'] = '2017'
df_2018['year'] = '2018'
df_2019['year'] = '2019'
df_2020['year'] = '2020'
df_2021['year'] = '2021'
df_2022['year'] = '2022'
df_2023['year'] = '2023'
```

#### 04. Data Cleaning and Consistency Check

```python
# Concatenating data
df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023])
print(df.shape)
df.head()
```

##### Mixed data types

```python
# Check data types
df.dtypes
```

```python
# Check for mixed types 
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Change 'year' data type
df['year'] = df['year'].astype(int)
```

##### Missing values

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop line
df = df.drop('State of Palestine')

# NOTE: Palestine apears only in 2023 so i will drop it
```

```python
mean = df.loc[df.index == 'United Arab Emirates', 'perceptions_of_corruption'].mean()
df.loc[(df.index == 'United Arab Emirates') & (df['year'] == 2018), 'perceptions_of_corruption'] = mean
df['corruption_perceptions'].fillna(mean, inplace=True)
```

```python
# Check for missing values
df.isnull().sum()
```

##### Duplicated values

```python
# Check for duplicated values
df_dups = df[df.duplicated()]
df_dups
```

```python
# Change column names for consistency
df = df.rename(columns = {'perceptions_of_corruption' : 'corruption_perceptions'})
```
